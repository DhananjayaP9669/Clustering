```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
crime=pd.read_csv('crime_data.csv')
crime
```

```python
from sklearn.preprocessing import MinMaxScaler
trans= MinMaxScaler()
data = pd.DataFrame(trans.fit_transform(crime.iloc[:,1:]))
data
```

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
z = linkage(data, method='complete',metric='euclidean')
plt.figure(figsize=(15,5))
plt.title('Heirarchical Clustering Dendogram')
plt.xlabel('Distance')
sch.dendrogram(z,)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
h_complete = AgglomerativeClustering(n_clusters=5, linkage='complete',affinity='euclidean').fit(data)
```

```python
clusters_labels=pd.Series(h_complete.labels_)
clusters_labels
```

```python
crime['clust']=clusters_labels
crime
```

```python
crime.iloc[:,1:].groupby(crime.clust).mean()
```

```python
crime.plot(x='Assault',y='Murder',kind='scatter',s=50,cmap=plt.cm.coolwarm)
plt.title('Clusters')
```
